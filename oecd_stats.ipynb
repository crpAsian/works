```python
import pandas as pd
```

```python
oecd_phys = pd.read_csv('./refs/HEALTH_REAC_11032024040207583.csv')
```

```python
oecd_phys.tail()
```

```python
oecd_phys.columns
```

```python
oecd_phys = oecd_phys[['Variable', 'Measure', 'Country', 'Year', 'Value', 'Flags']]
oecd_phys
```

```python
oecd_per = oecd_phys[(oecd_phys['Measure']=='Density per 1 000 population (head counts)') & (oecd_phys['Year'] >= 2020)
          & (oecd_phys['Variable']=='Practising physicians')]
```

```python
oecd_per['Country'].unique()
```

```python
non_oecd = ['Argentina', 'Brazil', 'Bulgaria', "China (People's Republic of)", 'Croatia',
            'Indonesia', 'Peru', 'Romania', 'South Africa']
```

```python
oecd_per = oecd_per[oecd_per['Country'].apply(lambda x: x not in non_oecd)]
oecd_per
```

```python
oecd_per.groupby(by='Year').count() # 일단 2020년으로 ㄱ
```

```python
per_20 = oecd_per[oecd_per['Year']==2020]
```

```python
per_20.sort_values(by='Value', ascending=False, axis=0)
```
